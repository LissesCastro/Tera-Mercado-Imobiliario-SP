```python
import warnings
warnings.filterwarnings("ignore")

#Bibliotecas de importação de dados
import basedosdados as bd

#Manipulação de dados
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df_censo_basico = bd.read_table(dataset_id='br_ibge_censo_demografico',
table_id='setor_censitario_basico_2010', billing_project_id="229299106527")

df_domicilio_caracteristicas = bd.read_table(dataset_id='br_ibge_censo_demografico',
table_id='setor_censitario_domicilio_caracteristicas_gerais_2010', billing_project_id="229299106527")

df_cor_raca = bd.read_table(dataset_id='br_ibge_censo_demografico',
table_id='setor_censitario_raca_idade_genero_2010', billing_project_id="229299106527")

df_filhos = bd.read_table(dataset_id='br_ibge_censo_demografico',
table_id='setor_censitario_idade_total_2010', billing_project_id="229299106527")

```

```python
estado = ['SP1', 'SP2']

for base in [df_censo_basico, df_domicilio_caracteristicas, df_cor_raca, df_filhos]:
    base = base[base['sigla_uf'].isin(estado)]
```

```python
df_censo_basico
```

```python
df_censo_basico = df_censo_basico.rename({'v001':'Domicílios_Setor', 'v003':'Média_Moradores_por_domicilio_setor', 'v005':'Média_Renda_Responsaveis_Setor'}, axis=1)
df_censo_basico = df_censo_basico[['id_setor_censitario', 'Domicílios_Setor', 'Média_Moradores_por_domicilio_setor', 'Média_Renda_Responsaveis_Setor']]
```

```python
df_censo_basico 
```

```python
df_domicilio_caracteristicas = df_domicilio_caracteristicas[['id_setor_censitario', 'v001', 'v008', 'v050', 'v051', 'v052', 'v053', 'v054', 'v055', 'v056', 'v057', 'v058']]
df_cor_raca = df_cor_raca[['id_setor_censitario', 'v001', 'v002', 'v003', 'v004', 'v005', 'v006']]
df_filhos = df_filhos[['id_setor_censitario', 'v005', 'v006', 'v007']]

df_domicilio_caracteristicas = df_domicilio_caracteristicas.rename({'v001':'Domicilios no setor censitário', 'v008':'Domicilios alugados', 'v050':'Domicilios unipessoais'}, axis=1)
df_domicilio_caracteristicas['Domicilios 2 ou 3 moradores'] = df_domicilio_caracteristicas['v051']+df_domicilio_caracteristicas['v052']
df_domicilio_caracteristicas['Domicilios 4 ou mais moradores'] = df_domicilio_caracteristicas['v053']+df_domicilio_caracteristicas['v054']+df_domicilio_caracteristicas['v055']+df_domicilio_caracteristicas['v056']
df_domicilio_caracteristicas = df_domicilio_caracteristicas.drop(['v051', 'v052', 'v053', 'v054', 'v055', 'v056', 'v057', 'v058'], axis=1)

df_filhos['Total de filhos dos responsáveis de domicílios'] = df_filhos['v005']+df_filhos['v006']+df_filhos['v007']
df_filhos.drop(['v005', 'v006', 'v007'], axis=1)
```

```python
df_cor_raca = df_cor_raca.rename({'v001': 'Residentes no setor censitário', 'v002': 'Brancos', 'v003': 'Pretos', 'v004': 'Amarelos', 'v005':'Pardos', 'v006':'Índigenas'}, axis=1)
df_domicilio_caracteristicas = df_domicilio_caracteristicas.rename({'v008':'Domicilios Alugados'}, axis=1)
```

```python
dfs = [df_censo_basico, df_domicilio_caracteristicas, df_cor_raca, df_filhos]
```

```python
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['id_setor_censitario'],
                                            how='outer'), dfs)
```

```python
df_merged.drop(['v005', 'v006', 'v007'], axis=1, inplace=True)
```

```python
df_merged.to_csv('CENSO_2010_TRABALHADO.csv')
```
